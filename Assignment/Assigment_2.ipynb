Q34. Group data by Department and calculate total Salary.

```python
data_q34 = {
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [60000, 75000, 62000, 80000, 70000]
}
df_q34 = pd.DataFrame(data_q34)

print("Original DataFrame:")
print(df_q34)

total_salary_by_department = df_q34.groupby('Department')['Salary'].sum()
print("\nTotal Salary by Department:")
print(total_salary_by_department)
```

Q36. Merge two DataFrames using a common key.

```python
data1 = {'ID': [1, 2, 3, 4], 'Name': ['Alice', 'Bob', 'Charlie', 'David']}
df1 = pd.DataFrame(data1)

data2 = {'ID': [1, 2, 3, 5], 'Score': [90, 85, 92, 78]}
df2 = pd.DataFrame(data2)

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

df_merged = pd.merge(df1, df2, on='ID')
print("\nMerged DataFrame (inner join by default on 'ID'):")
print(df_merged)
```

Q37. Perform a left join between two DataFrames.

```python
df_left_merged = pd.merge(df1, df2, on='ID', how='left')
print("DataFrame 1 (Left):")
print(df1)
print("\nDataFrame 2:")
print(df2)
print("\nLeft Merged DataFrame:")
print(df_left_merged)
```

Q38. Concatenate two DataFrames row-wise.

```python
data_concat1 = {'ID': [1, 2], 'Value': [10, 20]}
df_concat1 = pd.DataFrame(data_concat1)

data_concat2 = {'ID': [3, 4], 'Value': [30, 40]}
df_concat2 = pd.DataFrame(data_concat2)

print("DataFrame 1 for concatenation:")
print(df_concat1)
print("\nDataFrame 2 for concatenation:")
print(df_concat2)

df_concatenated_rowwise = pd.concat([df_concat1, df_concat2], axis=0)
print("\nRow-wise Concatenated DataFrame:")
print(df_concatenated_rowwise)
```

Q39. Concatenate two DataFrames column-wise.

```python
data_col_concat1 = {'A': [1, 2], 'B': [10, 20]}
df_col_concat1 = pd.DataFrame(data_col_concat1)

data_col_concat2 = {'C': [3, 4], 'D': [30, 40]}
df_col_concat2 = pd.DataFrame(data_col_concat2)

print("DataFrame 1 for column-wise concatenation:")
print(df_col_concat1)
print("\nDataFrame 2 for column-wise concatenation:")
print(df_col_concat2)

df_concatenated_colwise = pd.concat([df_col_concat1, df_col_concat2], axis=1)
print("\nColumn-wise Concatenated DataFrame:")
print(df_concatenated_colwise)
```

```python
# Q40. Mini Project: Create a sales DataFrame and answer business questions

# Create a sample sales DataFrame
sales_data = {
    'Date': pd.to_datetime(['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-03']),
    'Region': ['East', 'West', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'A'],
    'Units_Sold': [10, 5, 12, 8, 7, 15],
    'Price_Per_Unit': [2.5, 10.0, 2.5, 5.0, 10.0, 2.5]
}
df_sales = pd.DataFrame(sales_data)

# Add a 'Total_Revenue' column
df_sales['Total_Revenue'] = df_sales['Units_Sold'] * df_sales['Price_Per_Unit']

print("Sales DataFrame created:")
print(df_sales)
```

```python
# Business Question 1: Total Revenue
total_revenue = df_sales['Total_Revenue'].sum()
print(f"Total Revenue Generated: ${total_revenue:.2f}")
```

```python
# Business Question 2: Total Units Sold by Region
units_sold_by_region = df_sales.groupby('Region')['Units_Sold'].sum()
print("Total Units Sold by Region:")
print(units_sold_by_region)
```

```python
# Business Question 3: Total Revenue by Product
revenue_by_product = df_sales.groupby('Product')['Total_Revenue'].sum()
print("Total Revenue by Product:")
print(revenue_by_product)
```
