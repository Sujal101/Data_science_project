**Q12**. Explain why np.arange(10).reshape(3,4) produces an error.

The error in `np.arange(10).reshape(3,4)` occurs because `np.arange(10)` creates an array with 10 elements (0 to 9). When you try to reshape it into a 3x4 matrix, NumPy expects 3 * 4 = 12 elements. Since the number of elements in the original array (10) does not match the number of elements required for the new shape (12), a `ValueError` is raised.

```python
import numpy as np

# This will produce an error
try:
    error_array = np.arange(10).reshape(3, 4)
    print(error_array)
except ValueError as e:
    print(f"Error: {e}")
```

To correctly reshape an array with 10 elements, the product of the new dimensions must equal 10. For example, a 2x5 or 5x2 matrix would work:

```python
import numpy as np

correct_array = np.arange(10)

# Correctly reshape to 2x5
reshaped_2x5 = correct_array.reshape(2, 5)
print("Reshaped to 2x5:\n", reshaped_2x5)

# Correctly reshape to 5x2
reshaped_5x2 = correct_array.reshape(5, 2)
print("\nReshaped to 5x2:\n", reshaped_5x2)
```
